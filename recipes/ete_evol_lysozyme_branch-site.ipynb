# Reproduce codeml analysis from Yang Z (MBE, 2005)

This cookbook aims at reproducing the results about **selection on sites** presented in the work [Yang Z (MBE, 2005)](http://www.ncbi.nlm.nih.gov/pubmed/15689528) using `ete-evol`.
 
The data files adapted from the examples in the PAML package, large dataset will be used.

This files are:
* data/lysozyme/lysozymeLarge.fasta
* data/lysozyme/lysozymeLarge.nw

 

```python
%%bash
ete3 view -t data/lysozyme/lysozymeLarge.nw  --alg data/lysozyme/lysozymeLarge.fasta --alg_type compactseq \
-i data/lysozyme/lysozymeLarge.png
```

```python
from IPython.display import Image
Image(filename='data/lysozyme/lysozymeLarge.png')
```

In particular we aim at reproducing the highlighted part the Table2:

```python
from IPython.display import Image
Image(filename='data/lysozyme/Table2_YangZ_2002.png')
```

We are thus going to run 6 models.

First we run models M1 and M2:

```python
%%bash
ete3 evol -t data/lysozyme/lysozymeLarge.nw --alg data/lysozyme/lysozymeLarge.fasta --models M1 M2 -o /tmp/ete3-tmp/ \
--cpu 2
```

```python
%%bash
ete3 evol -t data/lysozyme/lysozymeLarge.nw --alg data/lysozyme/lysozymeLarge.fasta --models M3.k2 -o /tmp/ete3-tmp/ \
--cpu 1
```

```python
%%bash
ete3 evol -t data/lysozyme/lysozymeLarge.nw --alg data/lysozyme/lysozymeLarge.fasta --models M3.k3 -o /tmp/ete3-tmp/ \
--cpu 1 --codeml_param kappa,3
```

```python
%%bash
ete3 evol -t data/lysozyme/lysozymeLarge.nw --alg data/lysozyme/lysozymeLarge.fasta --models M7 M8 -o /tmp/ete3-tmp/ \
--cpu 2
```
